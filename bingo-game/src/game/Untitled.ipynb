```python
class Card:

    def __init__(self, name, x_lenght= 5, y_lenght= 3):
        self.name = name
        self.layout = np.zeros((x_lenght, y_lenght))
        self.card = None

    def numbers(self, rng=None):
        col1 = np.random.choice([i for i in range(1,16)], 5, replace=False)
        col2 = np.random.choice([i for i in range(16,31)], 5, replace=False)
        col3 = np.random.choice([i for i in range(31,46)], 5, replace=False)
        col4 = np.random.choice([i for i in range(46,61)], 5, replace=False)
        col5 = np.random.choice([i for i in range(61,76)], 5, replace=False)
        col1 = col1.reshape((5,1))
        col2 = col2.reshape((5,1))
        col3 = col3.reshape((5,1))
        col4 = col4.reshape((5,1))
        col5 = col5.reshape((5,1))
        card = np.concatenate((col1,col2,col3,col4,col5), axis=1)
        self.card = card
        return card
    
```

```python
luis = Card("luis")
```

```python
luis.numbers()
```
